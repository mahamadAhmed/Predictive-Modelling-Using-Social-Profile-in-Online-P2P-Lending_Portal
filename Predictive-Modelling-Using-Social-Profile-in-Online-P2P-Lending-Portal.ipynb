### handel missing value

```python
data['EmploymentStatus'].fillna('Other',inplace=True)
data['Occupation'].fillna('Other',inplace=True)
categorical_features = data.select_dtypes(include=['category','object']).columns
data[categorical_features] = data[categorical_features].fillna(data[categorical_features].mode().iloc[0])

num_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[num_cols] = data[num_cols].fillna(data[num_cols].mean())
data.isnull().sum()
```

### Feature Engineering

```python
# Identify categorical columns
cat_cols = data.select_dtypes(include=['category','object']).columns

# Identify numerical columns
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from scipy.sparse import hstack
from sklearn.decomposition import TruncatedSVD
# Create a one-hot encoder object and fit and transform the categorical columns
ohe = OneHotEncoder(sparse=True, handle_unknown='ignore')
cat_data = ohe.fit_transform(data[cat_cols])
print(f"Encoded data shape: {cat_data.shape}")
# Scale numerical columns
```

```python
new_columns = ohe.get_feature_names(input_features=cat_cols)
new_columns
```

```python
# Drop categorical columns not present in data
cat_cols = list(set(cat_cols) & set(data.columns))

# Apply scaling
scaler = StandardScaler(with_mean=False)
scaled_data = scaler.fit_transform(data.drop(columns=cat_cols).to_numpy())
# Apply normalization
normalizer = MinMaxScaler()
normalized_data = normalizer.fit_transform(data.drop(cat_cols, axis=1))

# Print the shape of the scaled data
print(f"Scaled data shape: {scaled_data.shape}")

# Print the shape of the normalized data
print(f"Normalized data shape: {normalized_data.shape}")
```

```python
num_cols = data.columns.difference(cat_cols)

scaler = MinMaxScaler()
num_data = scaler.fit_transform(data[num_cols])

# Combine one-hot encoded and scaled numerical data
data = hstack([num_data, cat_data])

# Apply Truncated SVD for feature extraction
svd = TruncatedSVD(n_components=50)
data = svd.fit_transform(data)

# Convert TruncatedSVD output to dense array
data = pd.DataFrame(data)

# Print the shape of the final data
print(f"Final data shape: {data.shape}")
```
